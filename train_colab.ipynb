```python
import matplotlib.pyplot as plt

#  
models = ['ResNet-152', 'EfficientNet-B1', 'ResNeXt-101', 'EfficientNet-B3',
          'SENet', 'NASNet-A', 'AmoebaNet-A', 'EfficientNet-B7']
params = [60, 7.8, 84, 12, 56, 89, 555, 66]
accuracy = [77.8, 79.1, 81.2, 81.6, 83.1, 82.7, 83.5, 84.3]

# Pareto frontier   (: EfficientNet )
pareto_params = [7.8, 12, 66]
pareto_acc = [79.1, 81.6, 84.3]

# 
plt.figure(figsize=(8, 6))
plt.grid(True)
plt.scatter(params, accuracy, color='black')

#   
for i, name in enumerate(models):
    if name == 'AmoebaNet-A':
        plt.text(params[i] - 60, accuracy[i] + 0.3, name, fontsize=8)
    else:
        plt.text(params[i] + 1, accuracy[i], name, fontsize=8)

# Pareto frontier 
plt.plot(pareto_params, pareto_acc, '-o', color='red')

plt.xlabel('Number of Parameters (Millions)')
plt.ylabel('Tiny ImageNet Top-1 Accuracy (%)')
plt.title('Model Trade-off: Accuracy vs Parameters')
plt.legend()
# plt.grid(False)
plt.tight_layout()
plt.show()
```
